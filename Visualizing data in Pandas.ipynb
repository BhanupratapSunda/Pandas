```python
df=df.set_index('Date')
df
```

```python
df.plot()             # By defalut it gives line Plot.
```

```python
df.plot(kind='line',title='Ice cream ratings', xlabel='Daily Ratings',ylabel='Scores')  
```

```python
df.plot(kind='bar')
```

```python
df.plot.scatter(x='Overall Rating',y='Flavor Rating')          # can be drawan between 2 columns of the Dataframe. We can't take index as x or y axis.
```

```python
df.plot.scatter(x='Overall Rating',y='Flavor Rating',s=100)      # s indicates the size of the dots. 
```

```python
df.plot.scatter(x='Overall Rating',y='Flavor Rating',s=100,c='Red')      # C indicates the Colour of the dots. 
```
